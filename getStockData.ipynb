```python
def appendToFile(filepath, content):
    f=open("data/{stockNo}.txt".format(stockNo=filepath), "a+")
    f.write(content)
    f.close()
    

def findStartIndex(lines):
    StartIndex = 0
    for line in lines:
        StartIndex = StartIndex + 1
        if "證券代號" in line:
            break
        else:
            continue
    return StartIndex
```

```python
from datetime import *
import requests
from random import randint
import time
#from time import sleep
#for i in range(10):
#    print ((datetime.now() + timedelta(days=i)).strftime("%Y%m%d"))

#startDate = datetime.strptime("20191120","%Y%m%d")
startDate = datetime.strptime("20190501","%Y%m%d")
endDate = datetime.strptime("20191121","%Y%m%d")
diffDays = (endDate - startDate).days + 1
#print((endDate - startDate).days)
skipEqualSign = 1
for curDate in range(diffDays):
    requestDateStr = (startDate + timedelta(days=curDate)).strftime("%Y%m%d")
    #print("post:"+'https://www.twse.com.tw/exchangeReport/MI_INDEX?response=csv&date=' + requestDateStr + '&type=ALL')
    # 下載股價
    print("starting:"+requestDateStr)
    sleepsec = randint(10,20)
    print("sleep:" + str(sleepsec))
    time.sleep(sleepsec)
    r = requests.post('https://www.twse.com.tw/exchangeReport/MI_INDEX?response=csv&date=' + requestDateStr + '&type=ALL')
    requestLines = r.text.split("\r\n")
    if r.text:
        startIndex = findStartIndex(requestLines)
        #print(startIndex)
        totalLens =  len(requestLines)-1 # skip last empty line
        for i in range(startIndex, totalLens):
            line = requestLines[i][skipEqualSign:]
            line = line[:len(line)-1] # skip rest ,
            stockNo = line.split(",")[0]
            if len(stockNo.replace("\"","")) == 4: #only 4 digit is what we want
                newLine = '"{date}",{restLine}\n'.format(date=requestDateStr,restLine=line)
                #print(newLine)
                appendToFile(stockNo.replace("\"",""), newLine)
    
    print("done:"+ requestDateStr)
    #startIndex = findStartIndex(requestLines)
    #print(startIndex)
print("All done")
```
